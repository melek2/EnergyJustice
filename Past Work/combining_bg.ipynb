```python
import pandas as pd
import shapefile
import os


def shapefile_to_dataframe(file_name):
    # Load the shapefile
    sf = shapefile.Reader(file_name)

    # Get the fields and records
    fields = [field[0] for field in sf.fields[1:]]  # Exclude the DeletionFlag field
    records = sf.records()

    # Create a list to store all the records
    all_records = []

    # Iterate through the records and append them to the list
    for record in records:
        all_records.append(dict(zip(fields, record)))

    # Create a DataFrame from the list of records
    df = pd.DataFrame(all_records)

    return df
```

```python
folder_path = "/Users/melek/Desktop/Research/Research_Winter_2024/BG_shapefiles"

# Get a list of all shapefiles in the folder
shapefiles = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.zip')]

# Create an empty DataFrame to store all the data
combined_df = pd.DataFrame()

# Iterate through each shapefile, convert to DataFrame, and concatenate
for shapefile_path in shapefiles:
    df = shapefile_to_dataframe(shapefile_path)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Display the combined DataFrame
display(combined_df)
```
